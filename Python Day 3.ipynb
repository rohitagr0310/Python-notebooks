Conditional Statements

```python
if {condition}:
  # code block
```

````python
num = -10

if num > 0:
  print("Positive")
````

````python
num = 10

if num > 0:
  print("Positive")
else:
  print("Negative")
````

````python
num = 7

if num % 2 == 0:
  print("Even")
else:
  print("Odd")
````

````python
a = 10
b = 20

if a > b:
  print("a is greater than b")
else:
  print("b is greater than a")
````

````python
num = 0

if num > 0:
  print()
  print()
  print("positive")
elif num < 0:
  print("negative")
else:
  print("zero")
````

````python
a = 10
b = 5

if (a > b):
  print("a is greater than b")

if a > b:
  print("a is greater than b")
````

Loops

```python
while condition:
  # code block
  # terminating condition
```

````python
num = 1

while num <= 10:
  print(num)
  num += 1
````

````python
num = 10

while num >= 1:
  print(num)
  num = num - 1
````

``` python
for value in sequence:
  #code block
```

````python
range(10)
````

````python
print(list(range(10)))
````

````python
for value in range(1, 11):
  print(value)
````

````python
word = "Python"

for letter in word:
  print(letter)
````

````python
fruits = ["apple", "Banana", "Mango"]

for fruit in fruits:
  print(fruit)
````
